```r
x<-c(50,60,70,80)
y<-c(170,172,174,176)
relation<-lm(y~x)
print(relation)
summary(relation)
```

```r
plot(x,y,col="blue", main="Linear Regression Example",abline(lm(y~x)), cex=1.3,pch=16,xlab="x", ylab="y")
```

```r
confint(relation, level=0.95)
```

```r
plot(fitted(relation), residuals(relation))
```

```r
predict(relation, data.frame(x=67))
```
